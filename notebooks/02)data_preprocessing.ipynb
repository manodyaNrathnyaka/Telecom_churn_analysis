Model Training and Evaluation

Rf model

```python
rf_model=RandomForestClassifier(
    n_estimators=300, #number of trees
    random_state=42, #for reproducibility
    max_depth=10,
    min_samples_split=5,
    min_samples_leaf=2
)

rf_model.fit(x_train_smote,y_train_smote)
```

```python
rf_pred_train=rf_model.predict(x_train_smote)
```

```python
rf_test_pred=rf_model.predict(X_test)
```

```python
def training_scores(y_train,y_pred):
    accuracy=round(accuracy_score(y_train,y_pred),3)
    precision=round(precision_score(y_train,y_pred),3)
    recall=round(recall_score(y_train,y_pred),3)
    f1=round(f1_score(y_train,y_pred),3)
    print(f"Training  scores:\n\tAccuracy={accuracy}\n\tPrecision={ precision}\n\tRecall={recall}\n\tF1_score={f1}")

def validating_scores(y_test_smote,y_pred):
    accuracy=round(accuracy_score(y_test_smote,y_pred),3)
    precision=round(precision_score(y_test_smote,y_pred),3)
    recall=round(recall_score(y_test_smote,y_pred),3)
    f1=round(f1_score(y_test_smote,y_pred),3)
    print(f"Validating  scores:\n\tAccuracy={accuracy}\n\tPrecision={ precision}\n\tRecall={recall}\n\tF1_score={f1}")
```

```python
from sklearn.metrics import precision_score,recall_score,f1_score
```

```python
training_scores(y_train_smote,rf_pred_train)
validating_scores(y_test,rf_test_pred)
```

Hyper Para_Tuning

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
param_dist = {
    'n_estimators': [100, 200, 500],
    'max_depth': [5, 10, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2']
}

rf_random = RandomizedSearchCV(
    estimator=rf_model,
    param_distributions=param_dist,
    n_iter=10,
    cv=3,
    verbose=2,
    random_state=42,
    n_jobs=-1
)
```

```python
rf_nodel_tuned=rf_random.fit(x_train_smote,y_train_smote)
```

```python
print("Best Parameters:",rf_nodel_tuned.best_params_)
```

```python
rf_model_after_tuning=RandomForestClassifier(
    n_estimators=200, #number of trees
    random_state=42, #for reproducibility
    max_depth=7,
    min_samples_split=10,
    min_samples_leaf=3,
    max_features='sqrt'
)

rf_model_after_tuning.fit(x_train_smote,y_train_smote)
```

```python
new_random_f_pred_train=rf_model_after_tuning.predict(x_train_smote)
new_random_f_pred_test=rf_model_after_tuning.predict(X_test)
```

```python
training_scores(y_train_smote,new_random_f_pred_train)
validating_scores(y_test,new_random_f_pred_test)
```

XGBOOST training & evaluation

```python
from xgboost import XGBClassifier

# Create the model
xgb_model = XGBClassifier(
    n_estimators=100,
    learning_rate=0.02,
    max_depth=4,
    random_state=42,
    eval_metric='logloss',
    subsample=0.6,
    colsample_bytree=0.6,
    reg_alpha = 0.1, # L1 regularization
    reg_lambda = 1   # L2 regularization
)

# Train the model
xgb_model.fit(x_train_smote, y_train_smote)

# Predict
xgb_pred_on_train = xgb_model.predict(x_train_smote)
xgb_pred_on_test=xgb_model.predict(X_test)
```

```python
training_scores(y_train_smote,xgb_pred_on_train)
validating_scores(y_test,xgb_pred_on_test)
```

```python
xgb = XGBClassifier(eval_metric='logloss', random_state=42)

param_grid = {
    'n_estimators': [100, 200, 500],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0],
    'gamma': [0, 1, 5]
}

xgb_random = RandomizedSearchCV(
    estimator=xgb,
    param_distributions=param_grid,
    n_iter=10,
    cv=3,
    verbose=2,
    random_state=42,
    n_jobs=-1
)

xgb_random.fit(x_train_smote, y_train_smote)

print("Best parameters:", xgb_random.best_params_)
```

```python
from sklearn.metrics import classification_report

print("Random Forest Accuracy:", accuracy_score(y_test,new_random_f_pred_test))
print("XGBoost Accuracy:", accuracy_score(y_test, xgb_pred_on_test))

print("\nRandom Forest Report:\n", classification_report(y_test,new_random_f_pred_test ))
print("\nXGBoost Report:\n", classification_report(y_test, xgb_pred_on_test))
```
